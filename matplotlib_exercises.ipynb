# 1. Use matplotlib to plot the following equation: y = x^2 âˆ’ x + 2
#### You'll need to write the code that generates the x and y points.
#### Add an anotation for the point 0, 0, the origin.

```python
import matplotlib.pyplot as plt
```

```python
x = range(-100, 100)
y = [(number **2) - number + 2 for number in x]
```

```python
plt.plot(x, y)
plt.annotate('origin', (0, 2), (-.5, 20), arrowprops = dict())
plt.show()
```

# 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. 
#### Be sure to include a legend and an appropriate title for the figure.

```python
x = range(0, 10)

y1 = [math.sqrt(number) for number in x]
y2 = [x **3 for x in x]
y3 = [2 **x for x in x]
y4 = [1 / (x + 1) for x in x]
```

```python
plt.plot(x, y1, color = 'r', label = 'y = âˆšx')
plt.plot(x, y2, color = 'g', label = 'y = x^3')
plt.plot(x, y3, color = 'b', label = 'y = 2^x')
plt.plot(x, y4, color = 'y', label = 'y = 1 / (x+1)')

plt.legend()
plt.grid(True)
plt.xlim(0, 25)
plt.ylim(-10, 100)

plt.show()
```
